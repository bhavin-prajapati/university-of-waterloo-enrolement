# Problem
We want to analyze our dataset (waterloo-enrolement-2024.csv) using Apache Spark's MLlib with a given set of paramters which include the following as lists:
- Fiscal Year: 2016/17, 2017/18, ..., 2023/24
- Career: Undergraduate, Graduate
- Study Year: 1, 2, 3, 4, N, D, M
- Program Level: Diploma, Bachelors, Masters, Doctoral
- Faculty: ENG, ARTS, ENV, HEA, MATH, SCI, THL
- Program Grouping: Accounting, Applied Economics, Philosophy, Engineering, etc.
- Campus: University of Waterloo, St. Jerome's University, Conrad Grebel University College
- Term Type: Spring term, Fall term, Winter term, etc.
- Coop Regular: Regular, Coop
- Attendance: Full-Time, Part-Time
- Visa: Canadian, Canadian Permanent Resident, International

The solution should allow configuration of some or all the parameters and use them to train a model to predict student enrolment (Student Headcounts).

## Configurable Parameters
Configure which parameter values to include when training the model. Set any parameter to `None` to include all values, or provide a list to filter to specific values.

```python
# ====================================================================
# CONFIGURABLE PARAMETERS
# Set any parameter to None to include ALL values for that column.
# Otherwise, provide a list of values to filter on.
# ====================================================================

config = {
    'Fiscal Year':       None,  # e.g. ['2022/23', '2023/24']
    'Career':            None,  # e.g. ['Undergraduate']
    'Study Year':        None,  # e.g. ['1', '2', '3', '4']
    'Program Level':     None,  # e.g. ['Bachelors', 'Masters']
    'Faculty (group)':   None,  # e.g. ['ENG', 'MATH']
    'Program Grouping':  None,  # e.g. ['Computer Science', 'Engineering']
    'Campus':            None,  # e.g. ['University of Waterloo']
    'Term Type':         None,  # e.g. ['Fall term', 'Winter term']
    'Coop Regular':      None,  # e.g. ['Coop']
    'Attendance':        None,  # e.g. ['Full-Time']
    'Visa Status':       None,  # e.g. ['Canadian', 'International']
}

# Apply filters based on configuration
filtered_df = df
for col_name, values in config.items():
    if values is not None:
        filtered_df = filtered_df.filter(col(col_name).isin(values))

print(f'Rows after filtering: {filtered_df.count()} (out of {df.count()})')
```

## Data Preprocessing
Encode all categorical features using StringIndexer and OneHotEncoder, then assemble them into a feature vector.

```python
# Define categorical feature columns (all columns except the target)
categorical_cols = [c for c in filtered_df.columns if c != 'Student Headcounts']

# Drop rows with null target
filtered_df = filtered_df.dropna(subset=['Student Headcounts'])

# Feature engineering: extract numeric year from 'Fiscal Year' (e.g. '2016/17' -> 2016)
# This lets models capture year-over-year enrollment trends as a continuous feature
filtered_df = filtered_df.withColumn('numeric_year', substring(col('Fiscal Year'), 1, 4).cast('double'))

# Build StringIndexer and OneHotEncoder stages for each categorical column
indexers = []
encoders = []
encoded_cols = []

for c in categorical_cols:
    idx_col = c + '_idx'
    ohe_col = c + '_ohe'
    indexers.append(StringIndexer(inputCol=c, outputCol=idx_col, handleInvalid='keep'))
    encoders.append(OneHotEncoder(inputCol=idx_col, outputCol=ohe_col))
    encoded_cols.append(ohe_col)

# Assemble all encoded features + numeric year into a single vector
assembler = VectorAssembler(inputCols=encoded_cols + ['numeric_year'], outputCol='features')

print(f'Categorical columns: {categorical_cols}')
print(f'Number of feature columns: {len(categorical_cols)} categorical + 1 numeric (numeric_year)')
```

## Train/Test Split

```python
# Rename target column and apply log1p transform to handle right-skewed distribution
# (mean ~14, stddev ~30, max 552 — highly skewed count data)
# log1p(x) = log(1+x) safely handles zero values; we'll inverse with expm1 after prediction
model_df = filtered_df.withColumn('label', log1p(col('Student Headcounts').cast('double')))

# Keep original label for evaluation after inverse-transform
model_df = model_df.withColumn('original_label', col('Student Headcounts').cast('double'))

# Split data: 80% train, 20% test
train_df, test_df = model_df.randomSplit([0.8, 0.2], seed=42)

print(f'Training set size: {train_df.count()}')
print(f'Test set size: {test_df.count()}')
```

## Model Training with Cross-Validation
We train three regression models using MLlib with CrossValidator-based hyperparameter tuning:
1. **Linear Regression** — tuning regParam and elasticNetParam
2. **Random Forest Regressor** — tuning numTrees and maxDepth
3. **Gradient-Boosted Tree Regressor** — tuning maxIter, maxDepth, and stepSize

Models are trained on log-transformed labels. Predictions are inverse-transformed with expm1() and clipped to a minimum of 1 (headcounts must be positive).

```python
# Define models and their hyperparameter grids for CrossValidator tuning
lr = LinearRegression(featuresCol='features', labelCol='label', maxIter=100)
lr_grid = ParamGridBuilder() \
    .addGrid(lr.regParam, [0.001, 0.01, 0.1]) \
    .addGrid(lr.elasticNetParam, [0.0, 0.5, 1.0]) \
    .build()

rf = RandomForestRegressor(featuresCol='features', labelCol='label', featureSubsetStrategy='sqrt', seed=42)
rf_grid = ParamGridBuilder() \
    .addGrid(rf.numTrees, [100, 200]) \
    .addGrid(rf.maxDepth, [10, 15, 20]) \
    .build()

# GBT grid: focused on the most impactful parameters.
# - maxIter=100 is sufficient for ~61k rows (200 adds time with marginal benefit)
# - maxDepth 5-8 is the sweet spot for GBT (deeper trees overfit and are much slower)
# - stepSize 0.05 vs 0.1 is the most impactful tuning knob for GBT quality
gbt = GBTRegressor(featuresCol='features', labelCol='label', subsamplingRate=0.8, seed=42)
gbt_grid = ParamGridBuilder() \
    .addGrid(gbt.maxIter, [100]) \
    .addGrid(gbt.maxDepth, [5, 8]) \
    .addGrid(gbt.stepSize, [0.05, 0.1]) \
    .build()

models = {
    'Linear Regression': (lr, lr_grid),
    'Random Forest': (rf, rf_grid),
    'Gradient-Boosted Trees': (gbt, gbt_grid),
}

# Build pipeline stages (shared across models)
preprocessing_stages = indexers + encoders + [assembler]

# Evaluator for CrossValidator (uses log-space RMSE for tuning)
cv_evaluator = RegressionEvaluator(labelCol='label', predictionCol='prediction', metricName='rmse')

# Evaluators for final metrics (on original scale after inverse-transform)
evaluator_rmse = RegressionEvaluator(labelCol='original_label', predictionCol='final_prediction', metricName='rmse')
evaluator_r2 = RegressionEvaluator(labelCol='original_label', predictionCol='final_prediction', metricName='r2')
evaluator_mae = RegressionEvaluator(labelCol='original_label', predictionCol='final_prediction', metricName='mae')

# Cache training data to avoid recomputing it for every CV fold
train_df.cache()
train_df.count()  # materialize the cache

# Train and evaluate each model with 3-fold cross-validation
results = {}
trained_models = {}

for name, (model, grid) in models.items():
    print(f'Training {name} with 3-fold CV...')
    pipeline = Pipeline(stages=preprocessing_stages + [model])

    cv = CrossValidator(
        estimator=pipeline,
        estimatorParamMaps=grid,
        evaluator=cv_evaluator,
        numFolds=3,
        parallelism=2,  # train multiple grid points in parallel
        seed=42
    )

    cv_model = cv.fit(train_df)
    predictions = cv_model.transform(test_df)

    # Inverse-transform predictions from log-space back to original scale
    # and clip to minimum of 1 (headcounts must be positive)
    predictions = predictions.withColumn(
        'final_prediction',
        greatest(expm1(col('prediction')), lit(1.0))
    )

    rmse = evaluator_rmse.evaluate(predictions)
    r2 = evaluator_r2.evaluate(predictions)
    mae = evaluator_mae.evaluate(predictions)

    results[name] = {'RMSE': rmse, 'R2': r2, 'MAE': mae}
    trained_models[name] = cv_model

    print(f'  RMSE: {rmse:.4f}')
    print(f'  R²:   {r2:.4f}')
    print(f'  MAE:  {mae:.4f}')
    print()

train_df.unpersist()
```

## Model Comparison

```python
# Compare models side by side
results_pd = pd.DataFrame(results).T
results_pd.index.name = 'Model'
print(results_pd.to_string())

# Plot comparison
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

metrics = ['RMSE', 'R2', 'MAE']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

for i, metric in enumerate(metrics):
    axes[i].bar(results_pd.index, results_pd[metric], color=colors[i])
    axes[i].set_title(metric)
    axes[i].set_ylabel(metric)
    axes[i].tick_params(axis='x', rotation=30)

plt.suptitle('Model Comparison', fontsize=14)
plt.tight_layout()
plt.show()
```

## Best Model - Predictions vs Actuals
Select the best model based on R² and visualize its predictions.

```python
# Select best model by R²
best_model_name = max(results, key=lambda k: results[k]['R2'])
best_model = trained_models[best_model_name]

print(f'Best model: {best_model_name}')
print(f'  R²:   {results[best_model_name]["R2"]:.4f}')
print(f'  RMSE: {results[best_model_name]["RMSE"]:.4f}')
print(f'  MAE:  {results[best_model_name]["MAE"]:.4f}')
```

## Save Trained Models
Save the best pipeline model for each algorithm (including preprocessing stages) to disk so they can be loaded later by an API.
A metadata JSON file is also written with evaluation metrics and the name of the best model.

```python
import os, json as _json, shutil

MODELS_DIR = 'models'

# Clean previous run
if os.path.exists(MODELS_DIR):
    shutil.rmtree(MODELS_DIR)
os.makedirs(MODELS_DIR, exist_ok=True)

# Save each model's best pipeline (from CrossValidator) to its own directory.
# cv_model.bestModel is a PipelineModel that includes all preprocessing +
# the trained regressor, so it can transform raw input data end-to-end.
saved_paths = {}
for name, cv_model in trained_models.items():
    safe_name = name.lower().replace(' ', '_').replace('-', '_')
    model_path = os.path.join(MODELS_DIR, safe_name)
    cv_model.bestModel.save(model_path)
    saved_paths[name] = model_path
    print(f'Saved {name} -> {model_path}')

# Write metadata (metrics + best model) so an API can discover which model to load
metadata = {
    'best_model': best_model_name,
    'models': {}
}
for name in results:
    safe_name = name.lower().replace(' ', '_').replace('-', '_')
    metadata['models'][name] = {
        'path': saved_paths[name],
        'RMSE': results[name]['RMSE'],
        'R2': results[name]['R2'],
        'MAE': results[name]['MAE'],
    }

metadata_path = os.path.join(MODELS_DIR, 'metadata.json')
with open(metadata_path, 'w') as f:
    _json.dump(metadata, f, indent=2)

print(f'\nMetadata written to {metadata_path}')
print(f'Best model: {best_model_name}')
```

```python
# Predictions vs Actuals scatter plot
best_predictions = best_model.transform(test_df)
# Inverse-transform predictions from log-space and clip to minimum of 1
best_predictions = best_predictions.withColumn(
    'final_prediction',
    greatest(expm1(col('prediction')), lit(1.0))
)
pred_pd = best_predictions.select('original_label', 'final_prediction').toPandas()

plt.figure(figsize=(10, 8))
plt.scatter(pred_pd['original_label'], pred_pd['final_prediction'], alpha=0.3, s=10)

# Perfect prediction line
max_val = max(pred_pd['original_label'].max(), pred_pd['final_prediction'].max())
plt.plot([0, max_val], [0, max_val], 'r--', linewidth=2, label='Perfect Prediction')

plt.xlabel('Actual Headcounts')
plt.ylabel('Predicted Headcounts')
plt.title(f'Predictions vs Actuals ({best_model_name})')
plt.legend()
plt.tight_layout()
plt.show()
```

## Feature Importance
Display feature importance for the Random Forest model.

```python
# Get feature importance from the Random Forest model
# trained_models stores CrossValidatorModel objects, so access .bestModel
# to get the underlying PipelineModel
rf_pipeline = trained_models['Random Forest'].bestModel
rf_model = rf_pipeline.stages[-1]  # Last stage is the regressor

# Map importance to encoded column groups
importances = rf_model.featureImportances.toArray()

# Aggregate importance by original categorical column
# Each OHE column group corresponds to one original categorical column
col_importance = {}
idx = 0
for i, c in enumerate(categorical_cols):
    # Get the OneHotEncoder output size from the trained model
    ohe_model = rf_pipeline.stages[len(categorical_cols) + i]  # encoder stage
    size = ohe_model.categorySizes[0] - 1  # OHE drops last category
    col_importance[c] = sum(importances[idx:idx+size])
    idx += size

# Add numeric_year importance (last feature in the vector)
if idx < len(importances):
    col_importance['numeric_year'] = sum(importances[idx:])

# Sort and plot
imp_pd = pd.Series(col_importance).sort_values(ascending=True)

plt.figure(figsize=(10, 6))
imp_pd.plot(kind='barh')
plt.xlabel('Feature Importance')
plt.title('Feature Importance by Column (Random Forest)')
plt.tight_layout()
plt.show()
```
